```python
import requests
from bs4 import BeautifulSoup
```

```python
#import url

url = 'https://www.discoverlosangeles.com/things-to-do/the-13-scariest-places-in-los-angeles'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
Venue = []
```

```python
# append the data "title" into the list
for row in soup.find_all(class_="title"):
    Venue.append(row.text)
```

```python
Venue
```

```python
Address = []
```

```python
# append the data "address" into the list
for row in soup.find_all(class_="dla-field-address"):
    Address.append(row.text)
```

```python
Address
```

```python
d = {'Venue': Venue, 'Full Address': Address}
```

```python
la_df = pd.DataFrame(d)
```

```python
la_df
```

```python
la_df = la_df.replace({'\n\n\n\n\n':''}, regex=True)
```

```python
la_df["Zip"] = la_df["Full Address"].str.split().str[-1]
```

```python
la_df
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import geocoder # to get coordinates
import folium 
```

```python
# function to get coordinates
def get_latlng(zipcode):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{},  :Los Angeles, California'.format(zipcode))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
coords = [get_latlng(zipcode) for zipcode in la_df["Full Address"].tolist()]
```

```python
coords
```

```python
df_coords = pd.DataFrame(coords, columns=['Latitude', 'Longitude'])
```

```python
# merge the coordinates into the original dataframe
la_df['Latitude'] = df_coords['Latitude']
la_df['Longitude'] = df_coords['Longitude']
```

```python
la_df
```

```python
la_df.to_csv('la_df.csv')
```

```python
# get the coordinates of LA
address = 'Los Angeles, California'

geolocator = Nominatim(user_agent="my-app")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Los Angeles, California {}, {}.'.format(latitude, longitude))
```

```python
# create map of LA using latitude and longitude values
map_la = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, Venue in zip(la_df['Latitude'], la_df['Longitude'], la_df['Venue']):
    label = '{}'.format(Venue)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_la)  
    
map_la
```
